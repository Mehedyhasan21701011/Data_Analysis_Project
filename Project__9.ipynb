```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/content/hotel_booking.csv')
data
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data.drop(['company'], axis=1, inplace=True)
```

```python
data.drop(['agent'],axis = 1,inplace = True)
```

```python
data.isnull().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
# Data types of the columns
data.dtypes
```

```python
# Change date object to Date format?
data['reservation_status_date'] = pd.to_datetime(data.reservation_status_date)
```

```python
data.describe(include = object)
```

```python
data.hotel.unique()
```

```python
data.describe()
```

```python
# Remove outlier
# here adr value of 75% is less than 121 and 25 % of value between 12- to 384 ,so we can take adr value to 150.
data = data[data['adr']<150]
data.describe()
```

```python
data.head(100)
```

```python
# data cleaning

canceled_percentance = data.is_canceled.value_counts(normalize = True)*100
canceled_percentance
```

```python
plt.figure(figsize = (5,5))
plt.title('Reservatioin status count')
plt.bar(['Not canceled','Canceled'],data['is_canceled'].value_counts(),edgecolor = 'k',width = 0.7)
plt.show()
```

```python
plt.figure(figsize = (10,5))
ax1 = sns.countplot(x = 'hotel',hue = 'is_canceled',data = data,palette = 'Blues')
legend_labels,_ = ax1.get_legend_handles_labels()
ax1.legend(bbox_to_anchor = (1,1))
plt.title('Reservation status in different hotels')
plt.xlabel('Hotel')
plt.ylabel('Number of reservations')
plt.show()
```
