```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data_set = pd.read_csv("/content/sample_data/file (2).csv")
data_set
```

```python
# Find the null value of the dataset and fill it with the mean value of the column

data_set.isnull().sum()

data_set['Cylinders'].fillna(data_set['Cylinders'].mean(), inplace=True)
data_set

# remove null value
data_set.dropna(inplace=True)
data_set
#

data_set.isnull().sum()
```

```python
# Check what are the different types of makes in our dataset. And what is the count (occurrence) of each make

data_set['Make'].unique()

data_set.Make.value_counts()
```

```python
# Show all the records where origin is asia or europe

data_set[(data_set["Origin"] == 'Asia') | (data_set["Origin"] == 'Europe')]
```

```python
# Instruction ( Removing unwanted records ) - Remove all the records (rows) where Weight is above 4000

data_set[~(data_set['Weight']>2400)]
```
